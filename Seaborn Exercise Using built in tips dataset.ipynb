## In this notebook we will practice Seaborn with built in dataset

```python
#import libraries for dataframe and array manipulations
import pandas as pd
import numpy as np
```

```python
#data visualization libraries 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Import dataset
tips = sns.load_dataset('tips')
```

```python
#view data imported header and footer
tips
```

```python
#plot distribution of total_bill
sns.distplot(tips['total_bill'])
```

```python
#add more bins, break down total_bill to 100 bins
sns.distplot((tips['total_bill']),kde=False,bins=100)
```

```python
#draw scatter plot and frequency between total_bill and tip
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
#a heatmap like view of same scatter plot
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
#find scatter plot between all numerical features
sns.pairplot(tips)
```

```python
#Add categorical feature
sns.pairplot(tips,hue='sex')
```

```python
#Comapre time of meal with smoker and size
sns.relplot(x="total_bill", y="tip", col="time",
            hue="smoker", style="smoker", size="size",
            data=tips)
```

```python
#another way for the same graph
sns.relplot(x="total_bill", y="tip", col="time",
            hue="size", style="sex", size="size",
            palette="YlGnBu", markers=["D", "o"], sizes=(10, 125),
            edgecolor=".2", linewidth=.5, alpha=.75,
            data=tips);
```

```python
#predict the uncertainty of linear regression models
sns.lmplot(x="total_bill", y="tip", 
           col="time", hue="sex",
           data=tips);
```

```python
#breakdown of total_bill by day and sex
sns.catplot(x="day", y="total_bill", hue="sex",
            kind="swarm", data=tips);
```

```python
#finding the density plot of each 
sns.catplot(x="day", y="total_bill", hue="sex",
            kind="violin", split=True, data=tips);
```

```python
#Find the confidence interval of the mean value of each day by sex
sns.catplot(x="day", y="total_bill", hue="sex",
            kind="bar", data=tips);
```

```python
#combine matplotlib with seaborn for multiple plots
f, axes = plt.subplots(1, 2, sharey=True, figsize=(6, 4))
sns.boxplot(x="day", y="tip", data=tips, ax=axes[0])
sns.scatterplot(x="total_bill", y="tip", hue="day", data=tips, ax=axes[1]);
```

```python
#find correlation in numeric data
sns.heatmap(tips.corr(),cmap='coolwarm',annot=True)
```
